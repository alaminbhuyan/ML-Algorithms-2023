```python
class NumericalLinearRegression:
    def __init__(self):
        self.coef_ = None # m
        self.intercept_ = None # c
    
    def fit(self, X_train, y_train):
        numerator = (np.mean(X_train) * np.mean(y_train)) - np.mean((X_train * y_train))
        denominator = (np.mean(X_train))**2 - np.mean(X_train**2)
        
        self.coef_ = numerator / denominator
        
        self.intercept_ = np.mean(y_train) - (self.coef_ * np.mean(X_train))    
    
    def predict(self, X_test):
        return round((self.coef_ * X_test + self.intercept_), 3)
    
    def score(self, X_test, y_test):
        numerator = 0
        denominator = 0
        
        for index_pos, value in enumerate(X_test):
            numerator += (model.predict(value) - np.mean(y_test))**2
            denominator += (y_test[index_pos] - np.mean(y_test))**2
            
        return round((numerator / denominator), 2)
```
